2. Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.


```python
from datetime import date
from dateutil.relativedelta import relativedelta

class person:

  def __init__(self,name,dob,country):
    self.name = name
    self.dob = dob
    self.country = country

  def age(self):

    age = relativedelta(date.today(),self.dob)
    return f"{age.years} years {age.months} months {age.days} days"
```

4. Write a Python program to create a class that represents a shape. Include methods to calculate its area and perimeter. Implement subclasses for different shapes like circle, triangle, and square.


```python
import math
class shape:

  def area(self):
    pass

  def perimeter(self):
    pass

class circle(shape):

  def __init__(self,r):
    self.r = r

  def area(self):
    return round(math.pi*self.r**2,2)

  def perimeter(self):
    return round(2*math.pi*self.r)

class triangle(shape):

  def __init__(self,a,b,c):
    self.a = a
    self.b = b
    self.c = c

  def perimeter(self):
    return self.a + self.b + self.c

  def area(self):
    s = self.perimeter()/2
    return round((s*(s-self.a)*(s-self.b)*(s-self.c))**0.5,2)

class square(shape):

  def __init__(self,a):
    self.a = a

  def area(self):
    return self.a**2

  def perimeter(self):
    return self.a*4
```

```python
c1 = circle(4)
t1 = triangle(2,3,4)
s1 = square(5)

print(c1.perimeter())
print(c1.area())
print(t1.perimeter())
print(t1.area())
print(s1.perimeter())
print(s1.area())

shapes = [circle(5), square(4), triangle(3,4,5)]

for s in shapes:
    print(s.area(),end=' ')
```

5. Write a Python program to create a class representing a binary search tree. Include methods for inserting and searching for elements in the binary tree.


```python
class BinarySearchTree:

  def __init__(self, value = None):
    self.value = value
    self.left = None
    self.right = None

  def insert(self,value):

    if self.value == None:
      self.value = value
      return

    if value < self.value:
      if self.left == None:
        self.left = BinarySearchTree(value)
      else:
        self.left.insert(value)

    elif value > self.value:
      if self.right == None:
        self.right = BinarySearchTree(value)
      else:
        self.right.insert(value)

  def search(self,value):

    if self.value == None:
      return False

    if self.value == value:
      return True

    elif value < self.value:
      if self.left == None:
        return False
      return self.left.search(value)

    elif value > self.value:
      if self.right == None:
        return False
      return self.right.search(value)
```

```python
bst = BinarySearchTree()

bst.insert(10)
bst.insert(5)
bst.insert(15)
bst.insert(2)
bst.insert(7)

print("Search 7:", bst.search(7))
print("Search 20:", bst.search(20))
```

7. Write a Python program to create a class representing a linked list data structure. Include methods for displaying linked list data, inserting and deleting nodes.


```python
class LLDS:

  def __init__(self,value = None):
    self.value = value
    self.next = None

  def insert(self,value):

    if self.value == None:
      self.value = value
      return

    temp = self
    while temp.next:
      temp = temp.next

    temp.next = LLDS(value)

  def display(self):

    temp = self

    if temp.value is None:
      print("Linked List is empty")
      return

    while temp:
      print(temp.value, end=" -> ")
      temp = temp.next
    print("None")

  def delete(self, value):

    if self.value is None:
      print("List is empty")
      return

    if self.value == value:
      if self.next:
        self.value = self.next.value
        self.next = self.next.next
      else:
        self.value = None
      return

    temp = self
    while temp.next:
      if temp.next.value == value:
        temp.next = temp.next.next
        return
      temp = temp.next

    print("Value not found")

```

```python
ll = LLDS()

ll.insert(10)
ll.insert(20)
ll.insert(30)

ll.display()

ll.delete(20)
ll.display()
```

8. Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price.


```python
class shoppingcart:

  def __init__(self):
    self.price = {"land": 1000000,"car": 2000000,"flat": 10000000,"mobile": 150000}
    self.cart = {}

    print("available items")
    for key,value in self.price.items():
      print("item:",key,"price:",value)

  def add_item(self, item, quantity):
    if item not in self.price:
      print("Item not available")
      return

    if item in self.cart:
      self.cart[item] += quantity
    else:
      self.cart[item] = quantity

  def remove_item(self,item):
    if item in self.cart:
      del self.cart[item]
      return
    else:
      print("Item not added to cart")

  def total_price(self):

    total = 0
    for item, quantity in self.cart.items():
      total += self.price[item]*quantity
    return total

  def view_cart(self):

    for item,quantity in self.cart.items():
      price = self.price[item]
      print("item:",item,"  quantity:",quantity,"  total price for item:",quantity*price)

```

10. Write a Python program to create a class representing a queue data structure. Include methods for enqueueing and dequeueing elements.


```python
class queue:

  def __init__(self):
    self.queue = []

  def enqueue(self,value):

    self.queue.append(value)

  def dequeue(self):

    if len(self.queue) == 0:
      return "Queue is empty"

    else:
      del self.queue[0]

  def view_queue(self):
    return self.queue
```

```python
q = queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
q.enqueue(40)

print(q.view_queue())

q.dequeue()

q.view_queue()
```

14. Write a program that uses datetime module within a class. Enter manufacturing date and expiry date of the product. The program must display the years, months and days that are left for expiry.


```python
from datetime import datetime

class Product:

  def __init__(self):
    self.mfg_date = None
    self.exp_date = None

  def set_mfg_date(self,date):

    self.mfg_date = datetime.strptime(date,"%d-%m-%Y")

  def set_exp_date(self,date):

    self.exp_date = datetime.strptime(date,"%d-%m-%Y")

  def time_for_expiry(self):

    t = self.exp_date - self.mfg_date

    days = t.days

    years = days // 365
    months = (days % 365) // 30
    rem_days = (days % 365) % 30

    print("Time Remaining:")
    print(years,"years",months,"months",rem_days,"days")

```

```python
apple = Product()

apple.set_mfg_date("25-08-2024")

apple.set_exp_date("04-03-2031")

apple.time_for_expiry()
```

17. Create a Bus child class that inherits from the Vehicle class. The default fare charge of any vehicle is seating capacity * 100. Note: The bus seating capacity is 50. so the final fare amount should be 5500. You need to override the fare() method of a Vehicle class in Bus class.


```python
class Vehicle:

  def __init__(self,capacity):
    self.capacity = capacity

  def fare(self):
    return self.capacity * 100

class Bus(Vehicle):

  def __init__(self):
    super().__init__(50)

  def fare(self):
    base_fare = super().fare()
    maintenance = base_fare*0.10
    return base_fare + maintenance

```

	Overtime amount = (overtime * (salary / 50))


```python
class Employee:

  def __init__(self,emp_name,emp_ID,emp_salary,emp_department):
    self.emp_name = emp_name
    self.emp_ID = emp_ID
    self.emp_salary = emp_salary
    self.emp_department = emp_department

  def emp_assign_department(self,department):
    self.emp_department = department
    return

  def calculate_emp_salary(self,hours):
    if hours > 50:
      overtime = hours - 50
      overtime_amount = overtime * (self.emp_salary/50)
      total_salary = self.emp_salary + overtime_amount

    else:
      total_salary = self.emp_salary

    return total_salary

  def print_employee_details(self):

    print(f"Employee name: {self.emp_name} \nEmployee ID: {self.emp_ID} \nDepartment: {self.emp_department} \nSalary: {self.emp_salary}")
```

21. Write a Python class Inventory with attributes like item_id, item_name, stock_count, and price, and methods like add_item, update_item, and check_item_details.

```python
class Inventory:

    def __init__(self):
        self.items = {}

    def add_item(self, item_id, item_name, stock_count, price):
        if item_id in self.items:
            print("Item already exists")
        else:
            self.items[item_id] = {
                "name": item_name,
                "stock": stock_count,
                "price": price
            }

    def update_item(self, item_id, stock_count=None, price=None):
        if item_id in self.items:
            if stock_count is not None:
                self.items[item_id]["stock"] = stock_count
            if price is not None:
                self.items[item_id]["price"] = price
        else:
            print("Item not found")

    def check_item_details(self, item_id):
        if item_id in self.items:
            item = self.items[item_id]
            print(f"ID: {item_id}")
            print(f"Name: {item['name']}")
            print(f"Stock: {item['stock']}")
            print(f"Price: {item['price']}")
        else:
            print("Item not found")
```

```python
inv = Inventory()

inv.add_item(101, "mobile", 10, 50000)
inv.add_item(102, "charger", 50, 500)

inv.update_item(101, stock_count=8, price=48000)

inv.check_item_details(101)
inv.check_item_details(102)
```
