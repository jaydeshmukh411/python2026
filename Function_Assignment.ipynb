1. Write a Python function to find the maximum of three numbers.


```python
def max_of_three(a=0,b=0,c=0):

  l = max(a,b,c)

  return l

max_of_three(1,2,3)
max_of_three(67,68,90)
```

2. Write a Python function to sum all the numbers in a list.

	Sample List : (8, 2, 3, 0, 7)
	Expected Output : 20


```python
def sum_of_list(lst):

  sum = 0

  for i in lst:
    sum += i

  return sum

sum_of_list((8,2,3,0,7))
```

5. Write a Python function to check whether a number falls within a given range.


```python
def fall_in_range(a,b,c):

  if a in range(b,c+1):
    return "Falls in given range"

  else:
    return "Does not fall in given range"

fall_in_range(27,20,30)

fall_in_range(2,20,30)
```

```python
def check_range(a,b,c):

  if b <= a <= c:
    return "Falls in given range"

  else:
    return "Does not fall in given range"

check_range(27,20,30)

check_range(2,20,30)
```

6. Write a Python function that accepts a string and counts the number of upper and lower case letters.
	Sample String : 'The quick Brow Fox'
	Expected Output :
	No. of Upper case characters : 3
	No. of Lower case Characters : 12

```python
def total_upper_lower(a):

  low = 0
  up = 0
  for i in a:
    if i.islower():
      low += 1
    elif i.isupper():
      up += 1

  print("No. of Upper case characters:",up)
  print("No. of Lower case characters:",low)

total_upper_lower("The quick Brow Fox")
```

7. Write a Python function that takes a list and returns a new list with distinct elements from the first list.

	Sample List : [1,2,3,3,3,3,4,5]
	Unique List : [1, 2, 3, 4, 5]

```python
def unique_list(lst):

  l = []
  for i in lst:
    if i not in l:
      l.append(i)
  return l

unique_list([1,2,3,3,3,3,4,5])
```

8. Write a Python function that checks whether a passed string is a palindrome or not.


```python
def palindrome(s):

  a = ''.join(i for i in s if i != ' ')
  for i in range(len(a)):
    if a[i] != a[(len(a)-1) - i]:
      print("not a palindrome")
      break
  else:
    print("palindrome")

palindrome("madam")
palindrome("nurses run")
```

```python
def is_palindrome(a):

  if a == a[::-1]:
    print("Is a palindrome")
  else:
    print("not a palindrome")

is_palindrome("abccba")
is_palindrome("nursesrun")
```

9. Write a Python program that accepts a hyphen-separated sequence of words as input and prints the words in a hyphen-separated sequence after sorting them alphabetically.

	Sample Items : green-red-yellow-black-white
	Expected Result : black-green-red-white-yellow


```python
def al_order(s):

  a = s.split("-")
  print("-".join(sorted(a)))

al_order("green-red-yellow-black-white")
```

19. Write a Python program to rearrange positive and negative numbers in a given array using Lambda.

	Original arrays:
	[-1, 2, -3, 5, 7, 8, 9, -10]
	Rearrange positive and negative numbers of the said array:
	[2, 5, 7, 8, 9, -10, -3, -1]

```python
L = [-1, 2, -3, 5, 7, 8, 9, -10]

pos = list(filter(lambda x: x > 0,L))
neg = list(filter(lambda x: x < 0,L))[::-1]

res = pos + neg
print("Original array:")
print(L)

print("\nRearranged array:")
print(res)
```

22. Write a Python program to check whether a given string contains a capital letter, a lower case letter, a number and a minimum length using lambda.

```python
check_string = lambda s: (
    any(map(lambda x: x.isupper(), s)) and
    any(map(lambda x: x.islower(), s)) and
    any(map(lambda x: x.isdigit(), s)) and
    len(s) >= 6
    )


s = input("Enter a string: ")

if check_string(s):
    print("Valid string")
else:
    print("Invalid string")
```
