4. Write a Python program to read last n lines of a file.


```python
def read_last_lines(filename,n):
  with open(filename,'r') as f:
    L = f.readlines()

    for i in L[-n:]:
      print(i,end = '')

read_last_lines('sample.txt',3)
```

6. Write a Python program to read a file line by line store it into a variable.


```python
def read_file_lines(filename):
    lines = []

    with open(filename, 'r') as file:
        for line in file:
            lines.append(line.strip())

    return lines

read_file_lines('sample.txt')
```

7. Write a python program to find the longest words.


```python
def longest_word(filename):
    with open(filename, 'r') as file:
        words = file.read().split()

    longest_word = max(words, key=len)
    return longest_word

longest_word('sample.txt')
```

8. Write a Python program to count the number of lines in a text file.


```python
def number_of_lines(filename):
  with open(filename,'r') as f:
    L = f.readlines()

    count = len(L)

    return count

number_of_lines('sample.txt')
```

9. Write a Python program to count the frequency of words in a file.


```python
def words_count(filename):
  count = {}

  with open(filename,'r') as f:

    for line in f:
      words = line.split()

      for word in words:
        word = word.lower()

        if word in count:
          count[word] += 1
        else:
          count[word] = 1
  return count

print(words_count('sample.txt'))
```

12. Write a Python program to copy the contents of a file to another file .


```python
def copy_file(file1,file2):
    with open(file1, 'r') as f:
        with open(file2, 'w') as g:
            for line in f:
                g.write(line)

copy_file('sample.txt','sample1.txt')
```

13. Write a Python program to read a random line from a file.


```python
import random

def random_line(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()

    print(random.choice(lines).strip())

random_line('sample.txt')
```

14. Write a Python program to assess if a file is closed or not.


```python
def check_file_closed(filename):
    f = open(filename, 'r')

    f.close()

    print("File check:", f.closed)

check_file_closed("sample.txt")
```

15.Write a Python program that takes a text file as input and returns the number of words of a given text file.

```python
def count_words(filename):
    with open(filename, 'r') as file:
        text = file.read()
        words = text.split()

    return len(words)
count_words('sample.txt')
```
